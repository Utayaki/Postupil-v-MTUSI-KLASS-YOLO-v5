```python
import torch
from IPython.display import Image 
import os 
import random
import shutil
from sklearn.model_selection import train_test_split
import xml.etree.ElementTree as ET
from xml.dom import minidom
from tqdm import tqdm
from PIL import Image, ImageDraw
import numpy as np
import matplotlib.pyplot as plt

random.seed(108)
```

```python
xml_file = "Dataset_To_Process/annotations.xml"
root = ET.parse(xml_file).getroot()
```

```python
# Классы нужно представить вот так
class_name_to_id_mapping = {"Под-квадратик в квадратике QR-кода": 0,
                           "Тело QR-кода": 1}

def convert_to_yolov5(info_dict):
    print_buffer = []
    
    # Для каждой коробки 
    for b in info_dict["bboxes"]:
        try:
            class_id = class_name_to_id_mapping[b["class"]]
        except KeyError:
            print("Неправильный класс - смерть ", class_name_to_id_mapping.keys())
        
        # Коробка
        b_center_x = (b["xmin"] + b["xmax"]) / 2 
        b_center_y = (b["ymin"] + b["ymax"]) / 2
        b_width    = (b["xmax"] - b["xmin"])
        b_height   = (b["ymax"] - b["ymin"])
        
        # Нормалим коробку
        image_w, image_h, image_c = info_dict["image_size"]  
        b_center_x /= image_w 
        b_center_y /= image_h 
        b_width    /= image_w 
        b_height   /= image_h 
        
        #Ну тут для отладки можно посмотреть все ли квадратики записались
        print_buffer.append("{} {:.3f} {:.3f} {:.3f} {:.3f}".format(class_id, b_center_x, b_center_y, b_width, b_height))
        
    save_file_name = "Dataset_To_Process/" + os.path.join("annotations", info_dict["filename"].replace("jpg", "txt"))
    
    print("\n".join(print_buffer), file=open(save_file_name, "w"))
    
```

```python
for elem in root:
    if elem.tag == "image":
        data_for_one_picture = {'bboxes': [], "filename": elem.attrib["name"], 'image_size': (1280, 720, 3)}
        for elemm in elem:
            box = {"class": elemm.attrib['label'], 'xmin': float(elemm.attrib['xtl']),
                   'ymin': float(elemm.attrib['ytl']), 'xmax': float(elemm.attrib['xbr']),
                   'ymax': float(elemm.attrib['ybr'])}

            data_for_one_picture['bboxes'].append(box)
            
        if not len(data_for_one_picture["bboxes"]) == 0:
            print(elem.attrib["name"])
            convert_to_yolov5(data_for_one_picture)
```
